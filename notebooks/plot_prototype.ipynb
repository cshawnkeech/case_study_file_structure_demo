# Plot prototypes


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Randomly generate data

- [`numpy.random.default_rng`](https://numpy.org/doc/stable/reference/random/generator.html#numpy.random.default_rng)


```python
seed = 42

rng = np.random.default_rng(seed=seed)
```

```python
norm = rng.normal(
    loc=0, scale=1,
    size=10_000)
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
sns.kdeplot(norm, ax=ax)
fig.suptitle("Normal Distribution (10,000 samples)")

# plt.savefig('../img/norm_demo.png')
plt.show()
```
